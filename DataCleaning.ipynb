```python
import pandas as pd
import numpy as np
import numpy as np
```

```python
data=pd.read_csv("train.csv")

'''
Data Cleaning

1. Removing duplicate values
2. Replacing values filled "NaN" with 0 or mean value
3. Finding mean
4. Creating new dataframe
'''

print(data.head())
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
data.drop_duplicates(subset=None,inplace=True)
data.info()
```

```python
data=data.dropna()
data
```

```python
for i in range(len(data["Year"])):
    if data.iloc[i]["Year"]>np.int64(2023) or data.iloc[i]["Year"]<np.int64(2007):
        data=data.drop(data.index[i])
```

```python
data
```

```python
data["Mileage"]=data["Mileage"].str.replace("km/kg","")
data["Mileage"]=data["Mileage"].str.replace("kmpl","")
data["Engine"]=data["Engine"].str.replace("CC","")
data["Power"]=data["Power"].str.replace("bhp","")
data["New_Price"]=data["New_Price"].str.replace("Lakh","")
```

```python
for i in range(len(data["New_Price"])):
    a=data.iloc[i]["New_Price"].split(" ")
    if a[1]=="Cr":
        data=data.drop(data.index[i])
```

```python
data
```

```python
data["Seats"]=data["Seats"].astype(np.int64)
data["Power"]=data["Power"].astype(np.float64)
data["New_Price"]=data["New_Price"].astype(np.float64)
data["Engine"]=data["Engine"].astype(np.float64)
data["Mileage"]=data["Mileage"].astype(np.float64)
```

```python
data.drop(data.columns[data.columns.str.contains('unnamed', case=False)], axis=1, inplace=True)
data
```

```python
data["New_Price"]=data["New_Price"]*100000
data["Price"]=data["Price"]*100000
data["New_Price"]=data["New_Price"].astype(np.int64)
data["Price"]=data["Price"].astype(np.int64)
```

```python
data
```

```python
data.to_csv("Car_price_data_cleaned.csv",index=False)
```
